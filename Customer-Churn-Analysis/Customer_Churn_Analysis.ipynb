extremely difficult to recover.

**This analysis answers three core business questions:**
- Who is most likely to churn?
- What factors are driving customers to leave?
- What can the business do to stop it?

**Who benefits:** Customer Success Team, Executive Leadership, 
Sales & Marketing

## Section 2: Exploratory Data Analysis (EDA)
Before building any model we explore the data to understand 
distributions, patterns, and potential issues.

```python
# Churn distribution
print("Churn counts:")
print(df['Churn'].value_counts())
print()
print("Churn percentage:")
print(df['Churn'].value_counts(normalize=True).round(2) * 100)
```

**Interpretation:** 26% of customers churned — 1 in 4 customers 
is leaving. At an average monthly charge of $65, this represents 
approximately $1.5M in lost monthly recurring revenue.

```python
# Statistical summary
print(df.describe())
```

**Interpretation:**
- Average tenure: 32 months (~2.7 years)
- Monthly charges range: $18 to $118
